```python
from dotenv import load_dotenv
import os
import requests

load_dotenv()
session = os.getenv('session')

# Replace 'your_session_cookie_here' with your actual session cookie
cookie = {'session': session}

# Replace with the specific day and year of the puzzle
url = 'https://adventofcode.com/2023/day/3/input'

response = requests.get(url, cookies=cookie)

if response.status_code != 200:
    print(f"Error: {response.status_code}")   
```

```python
engine_schematic = response.text.split('\n')[:-1]

def has_adjacent_symbol(x, y, schematic):
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            if 0 <= x + dx < len(schematic) and 0 <= y + dy < len(schematic[0]):
                if not schematic[x + dx][y + dy].isdigit() and schematic[x + dx][y + dy] != '.':
                    return True
    return False

def sum_part_numbers(schematic):
    ans = 0
    number = ''
    positions = []
    for i in range(len(schematic)):
        for j in range(len(schematic[i])):
            char = schematic[i][j]
            if char.isdigit():
                number += char
                positions.append(j)
            if not char.isdigit() or j == len(schematic[i]) - 1:
                if number and any(has_adjacent_symbol(i, k, schematic) for k in positions):
                    ans += int(number)
                number = ''
                positions = []
    return ans

sum_part_numbers(engine_schematic)
```

```python
def has_adjacent_symbol(x, y, schematic):
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            if 0 <= x + dx < len(schematic) and 0 <= y + dy < len(schematic[0]):
                if schematic[x + dx][y + dy] == '*':
                    return (x + dx, y + dy)
    return ()

def sum_part_numbers(schematic):
    number = ''
    positions = []
    asterisk = {}
    for i in range(len(schematic)):
        for j in range(len(schematic[i])):
            char = schematic[i][j]
            if char.isdigit():
                number += char
                positions.append(j)
            if not char.isdigit() or j == len(schematic[i]) - 1:
                seen = set()
                for k in positions:
                    ouput = has_adjacent_symbol(i, k, schematic)
                    if len(ouput) > 0:
                        seen.add(ouput)
                for o in seen:
                    temp = asterisk.get(o, [])
                    temp.append(int(number))
                    asterisk[o] = temp
                number = ''
                positions = []
    return asterisk

def sum_gear(dictionery):
    ans = 0
    for values in dictionery.values():
        if len(values) == 2:
            ans += (values[0] * values[1])
    return ans

sum_gear(sum_part_numbers(engine_schematic))
```
